##### Problem 2.2
----

$R_1 = 1, R_2 = 2$. Plot $f_m(k)$ vs $k\in [0,20]$ for $m=0,1,2$.

```julia
r1 = 1;
r2 = 2;

using SpecialFunctions

# function declarations
function jm_(m::Int64,x::Float64)
    return 0.5 .* (SpecialFunctions.besselj(m-1,x) - SpecialFunctions.besselj(m+1,x))
end

function ym_(m::Int64,x::Float64)
    return 0.5 .* (SpecialFunctions.bessely(m-1,x)  - SpecialFunctions.bessely(m+1,x))
end

function f(m::Int64,k::Float64,r1::Int64,r2::Int64)
    return jm_(m,k.*r1).*ym_(m,k.*r2) - jm_(m,k.*r2).*ym_(m,k.*r1)
end

k = Array(0:0.01:20);

b = zeros(size(k));

using PyPlot

for m in 0:2
    for i in eachindex(b)
        b[i] = f(m,k[i],r1,r2);
    end
    plot(k,b)
end
ylim([-0.5,0.5])
xlabel(L"k")
ylabel(L"f_m(k)")
legend(["m=0","m=1","m=2"])
```

##### Problem 2.3

For $m=0$, find the first three (smallest $k>0$) solutions to $f_0(k)=0$.

```julia
k = Array(3:0.01:10);
b = zeros(size(k));

for i in eachindex(b)
    b[i] = f(0,k[i],r1,r2);
end

z = zeros(size(k));

plot(k,z);
plot(k,b);
xlabel(L"k");
ylabel(L"f_0(k)");
```

Initial guesses should be 3.2, 6.3, and 9.5

```julia
# functions for newton's method

function f0(k::Float64)
    return f(0,k,1,2)
end

# center-difference approx
function dirivative_est(func,x::Float64,dx::Float64)
    return (func(x+dx) - func(x-dx)) ./ (2.0 .* dx)
end

# zero-finding function
function getzero(x::Float64,y::Float64,m::Float64)
    return x - (y./m)
end

# newton's method function
function newton(func, x::Float64, tol::Float64, dx::Float64)
    if abs(func(x)) <= tol
        return x
    end
    
    newton(func, getzero(x,func(x),dirivative_est(func,x,dx)),tol,dx)
end 
```

Find the zeros for each point using the above functions

```julia
init_guess = [3.2,6.3,9.5];
true_zeros = zeros(3);

for i in eachindex(init_guess)
    true_zeros[i] = newton(f0,init_guess[i],1e-15,1e-9);
    println("solution k$(i): $(true_zeros[i])");
end
```

##### Problem 2.4

Verify that these solutions are orthogonal under inner product $\langle u,v\rangle = \int_{\Omega}\overline{u}v$.

```julia
# functions for numerical integration

# trapazoidal area function
function traparea(y1::Float64,y2::Float64,dx::Float64)
    return 0.5 * (y1 + y2) * dx
end

# trapazoidal integration approx function
function trapint(func,a::Float64,b::Float64,dx::Float64,sum::Float64)
    if a >= b
        return sum
    end
    sum += traparea(func(a),func(a+dx),dx)
    
    trapint(func,a+dx,b,dx,sum)
end

# numeric integration using trapazoidal approx
function integrate(func,a::Float64,b::Float64,dx::Float64)
    return trapint(func,a,b,dx,0.0)
end
 
```

Assume $\alpha=1$, and evaluate the inner product for each solution pair to ensure it is zero.

```julia
# inner product calculation functions

# function u(r)
function u(r::Float64,k::Float64,alpha::Float64,beta::Float64)
    return alpha*SpecialFunctions.besselj(0,k*r) + beta*SpecialFunctions.bessely(0,k*r)
end

function beta(alpha::Float64,k::Float64,r1::Float64)
    return alpha * jm_(0,k*r1) / ym_(0,k*r1)
end


#eval inner product for each pair of solutions, with dx of 1e-6
for k1 in true_zeros
    for k2 in true_zeros
        if k1 < k2
            beta1 = beta(1.0,k1,1.0);
            beta2 = beta(1.0,k2,1.0);
            ipfunc(r) = r*u(r,k1,1.0,beta1) * r*u(r,k2,1.0,beta2);
            # numerical integration from r1=1 to r2=2 with dx = 1e-6
            ip = integrate(ipfunc,1.0,2.0,1e-3);
            # check if zero 
            if ip <= 1e-15
                println("$(k1) and $(k2): orthogonal");
            else
                println("$(k1) abd $(k2): $(ip) (not orthogonal)");
            end
        end
    end
end
```
