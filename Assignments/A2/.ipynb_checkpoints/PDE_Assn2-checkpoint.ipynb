### Assignment 2

##### Problem 2.3
Initialize values.

```julia
L = 1; M = 100;
dx = L / (M+1);
x = dx*0.5:dx:L; # sequence from dx/2 to L with a step size of dx
```

Define $c(x)$ function, and the D and C matrices.

```julia
using LinearAlgebra

c(X) = exp.(3*X);

C = diagm(0 => c(x)); # C Matrix diagonal matrix where values are c(x)

diff1(m) = [ 1 zeros(1,m-1); diagm(0 => -1*ones(m)) + diagm(1 => ones(m-1))]; # Differential Matrix

D = diff1(M);
```

Calculate $A$ by $A = -D^TCD$ and find the eigenvalues and eigenvectors.

```julia
A = -D'*C*D;

lambda,U = eigen(A);
```

Plot the four smallest eigenvectors

```julia
using PyPlot

plot(dx:dx:L-dx, U[:,end-3:end]);
xlabel("x"); ylabel("eigenfunctions");
legend(["4th","3rd","2nd","1st"]);
```

Verify that the first two eigenfunctions are orthogonal.

```julia
abs(dot(U[:,end],U[:,end-1])) < 1e-15 # if True, then the dot product 0 in practice.
```

Verify second-order convergence in the eigenvalues.

```julia
lambdas = zeros(4);
idx = 1;
for m = [100 200 400 800] # iterate through difference M sizes, doubling each time
    D = diff1(m);         # calculate D matrix for given m size
    dx = L/(m+1);         # calculate dx for given m
    x = dx*0.5:dx:L;      # get x sequence
    C = diagm(0 => c(x)); # get C matrix
    A = -D'*C*D;          # calculate A
    l,U = eigen(A);       # find eigenvalues
    lambdas[idx] = l[end];# save smallest magnitude eigenval
    idx += 1;
end
print("Show the decrease factor from a doubling of resolution\n")
println(abs(lambdas[2]-lambdas[1]) /abs(lambdas[3]-lambdas[2]))
println(abs(lambdas[3]-lambdas[2]) /abs(lambdas[4]-lambdas[3]))
```

A doubling of resolution results in a quadruple decrease in difference, thus the eigenvalues are second-order convergent.
